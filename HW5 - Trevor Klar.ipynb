```python
#  My function to numerically differentiate a function. Takes in a list of x-values and y-values, and outputs 
# a list of y'-values. 
def ddiff(x,y):
    result = []
    for i in indeces(x):
        if i == len(x)-1:                          #  Since the method of divided differences returns one less element
            result.append(result[-1])              # than it is given, we just duplicate the last element here. 
            continue
        result.append((y[i+1]-y[i])/(x[i+1]-x[i])) #  If not on the last element, calculates a y' value and adds it
    return result                                  # to the list. 

def plotme(f,a,b):
    x = np.linspace(a,b,500)                        #--
    y = f(x)                                        # Calculates x-values, y-values, derivatives.
    y1 = ddiff(x,y)                                 # |
    y2 = ddiff(x,y1)                                #--
    plt.plot(x[:498],y[:498], label='$f(x)$');      #  Plots the 3 functions, and gives them labels. Since our ddiff() 
    plt.plot(x[:498],y1[:498], label='$f\'(x)$');   # function duplicates the last element, using it twice in a row  
    plt.plot(x[:498],y2[:498], label='$f\'\'(x)$'); # will always give a trailing 0 on the second derivative. To keep 
                                                    # the picture pretty, we just didn't plot the last element on all
                                                    # 3 graphs. 
    plt.legend()                                    #  Adds a legend to the plot. 
    return plt.axis()                               #  Returns a plt.axis() instance so we can change any of the
                                                    # options we want, after calling plotme(). 
```

```python
def poly(x): return 4*x**3-3*x**2+2
plotme(poly,0,1);
```

```python
%%latex
$$f(x) \approx \frac{a_0}{2} + \sum_{k=1}^{n} \left[ a_k \cos \left( \frac{2\pi k x}{P} \right) + b_k \sin \left( \frac{2\pi k x}{P} \right) \right]$$
```

```python
%%latex
$$a_0 = \frac{2}{P} \int_0^P f(x)dx$$
$$a_k = \frac{2}{P} \int_0^P f(x) \cos \left( \frac{2\pi k x}{P} \right) dx , \quad k \geq 1$$
$$b_k = \frac{2}{P} \int_0^P f(x) \sin \left( \frac{2\pi k x}{P} \right) dx$$
```

```python
def FourierA(f,P,k):
    def fcos(x): # This inner function was really the key to making this work. It took me a long time to 
        return f(x)*np.cos((2*np.pi*k*x)/P) # figure out that you can do this. I had just assumed that all 'def' 
    if k == 0:                              # statements had to be in the top level of code. 
        return (2/P)*scipy.integrate.quad(f,0,P)[0]
    else:
        return (2/P)*scipy.integrate.quad(fcos,0,P)[0]
    
def FourierB(f,P,k):
    def fsin(x):
        return f(x)*np.sin((2*np.pi*k*x)/P)
    return (2/P)*scipy.integrate.quad(fsin,0,P)[0]

def FourierSeries(f, P, x, n):
    P = float(P)
    n += 1
    a = [0] * n
    b = [0] * n
    for i in range(n):             # Calculates all the coefficients up front, to make the code easier to read.
        a[i] = FourierA(f, P, i)   # I'd like to have some kind of a way to save these coefficients, to improve 
        b[i] = FourierB(f, P, i)   # performance when calculating many x-values with the same other parameters, 
    result = a[0]/2                # but I'm not sure how to do that. 
    for k in irange(1,n-1):
        result += a[k]*np.cos((2*np.pi*k*x)/P) + b[k]*np.sin((2*np.pi*k*x)/P)
    return result
```

```python
# I wrote this function to make the ploting easier, since I knew I'd be doing it repeatedly.
def FourierPlot(func, P, n):
    def ShiftedFunc(x): return func(x-2*P) # The Fourier Series we're using only works from 0 to P, so in order to 
    # calculate it from -2P to 2P, I had to translate the function right by 2P. This way I can calculate y-values from
    # 0 to 4P, and just plot them using x-values from the original function (from -2P to 2P).
    
    x = np.linspace(-2*P,2*P,400)   # List of x-values, y-values, and Fourier Series y-values.
    y = []                          #
    for i in x:                     #
        y.append(func(i))           #
    fouriery = []                   #
    for i in x:                     #
        fouriery.append(FourierSeries(ShiftedFunc, (4*P), i+(2*P), n))
        
    x = x[200:299]                  # The problem says to only graph the caclulated values on the range (0, P), so 
    y = y[200:299]                  # here I throw out the rest of the values. I think we did all this extra 
    fouriery = fouriery[200:299]    # calculating because a Fourier series is less accurate near the edges of the range
                                    # you're calculating on. 
    
    plt.plot(x,y)                   # Here, we plot f(x) and the Fourier Series for it. 
    plt.plot(x,fouriery, label='$n = '+str(n)+'$')
    plt.xlim(0,P)
    #plt.axis("equal");
    return plt.axis()               # So we can make further modifications to the plot.
```

```python
def sawtooth(x): return 1-x
```

```python
FourierPlot(sawtooth, 1, 5);
FourierPlot(sawtooth, 1, 10);
FourierPlot(sawtooth, 1, 25);
FourierPlot(sawtooth, 1, 50);
plt.legend(loc="upper right",frameon=True);
```

```python
def piecewise(x): 
    P = 1
    if x < float(P)/2:
        return 1.0 
    elif x >= float(P)/2:
        return 0.0
```

```python
FourierPlot(piecewise, 1, 5);
FourierPlot(piecewise, 1, 10);
FourierPlot(piecewise, 1, 25);
FourierPlot(piecewise, 1, 50);
plt.legend(loc="upper right",frameon=True);
```
